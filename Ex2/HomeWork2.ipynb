```python
n = np.arange(0,100)
x = np.arange(0,1000)
cosini = np.cos(2*np.pi*0.1*n)
z1 = np.array(np.zeros(500))
z2 = np.array(np.zeros(400))
y = np.concatenate((z1,cosini,z2))
y_n = y+np.sqrt(0.5)*np.random.randn(y.size)
h = np.cos(2*np.pi*0.1*n) #wrong
h1 = np.exp(-2j*np.pi*0.1*n) #right
yc1 = np.convolve(h,y_n,'same')
yca = np.abs(np.convolve(h1,y_n,'same'))
fig,ax = plt.subplots(4,1)
ax[0].plot(y)
ax[1].plot(y_n)
ax[2].plot(yc1)
ax[3].plot(yca)
```
